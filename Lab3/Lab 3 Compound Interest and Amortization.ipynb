```python
# Dana Hughes ISC 3222
# Lab 3: Compound Interest and Amortization

import numpy as np
import matplotlib.pyplot as plt

# text formatting
class color:
    BOLD = '\033[1m'
    END = '\033[0m'

# initial values
x = 1000 # principle
i = 0.05 # annual interest
n = np.arange(0, 11, 1) # years

# define function that takes in x, i, and n, and computes the total amount
def compoundInterest(x, i, n):
    return x * (1 + i)**n
a = compoundInterest(x, i, n)

# plotting
plt.figure()
plt.title("Compound Interest")
plt.xlabel("Years")
plt.ylabel("US Dollars ($)")
plt.plot(n, a, "-o")
plt.show()

# display amount earned per year
print(color.BOLD + "Savings throughout the years:" + color.END)
for m, a in enumerate(compoundInterest(x, i, n), start = 0):
    print("Year:", m, "Amount: $", "{:.2f}".format(a))
```
